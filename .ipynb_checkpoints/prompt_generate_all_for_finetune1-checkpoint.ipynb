```python
new_df1 = pd.DataFrame({
    'Job Title': df['Job Title'],
    'text': df['Job Description'],
})
new_df2 = pd.DataFrame({
    'Job Title': df['Job Title'],
    'text': df['How to become'],
})
new_df3 = pd.DataFrame({
    'Job Title': df['Job Title'],
    'text': df['What it takes'],
})
new_df4 = pd.DataFrame({
    'Job Title': df['Job Title'],
    'text': df['Career path and progression'],
})

# Merge new_df1 and new_df2
merged_df1 = pd.concat([new_df1, new_df2])
merged_df2 = pd.concat([new_df3, new_df4])
# Reset the index
merged_df1.reset_index(drop=True, inplace=True)
merged_df2.reset_index(drop=True, inplace=True)
merged_df3 = pd.concat([merged_df1, merged_df2])
merged_df3.reset_index(drop=True, inplace=True)
# Display the merged DataFrame
print(merged_df3)
```

```python
merged_df3.drop_duplicates(inplace=True)

# Reset the index
merged_df3.reset_index(drop=True, inplace=True)
merged_df3
```

```python
# Create a new DataFrame to store the updated rows
new_data = []

# Loop through each row in 'merged_df3'
for _, row in merged_df3.iterrows():
    job_title = row['Job Title']
    text = row['text']
    
    # Check if the value is a non-null string
    if isinstance(text, str):
        # Split the text into sentences using '.'
        sentences = text.split('.')
        
        # Remove any leading or trailing whitespaces and filter out empty sentences
        sentences = [sentence.strip() for sentence in sentences if sentence.strip()]
        
        # Create new rows with each sentence and the corresponding 'Job Title'
        new_rows = [{'Job Title': job_title, 'text': sentence} for sentence in sentences]
        
        # Extend the new data list with the new rows
        new_data.extend(new_rows)

# Create the new DataFrame
new_df = pd.DataFrame(new_data)

# Display the updated DataFrame
new_df
```

```python
# Function to remove punctuation and special characters from a string
def remove_special_characters(text):
    # Define the regular expression pattern to match punctuation and special characters
    pattern = r'[^a-zA-Z0-9\s]'
    
    # Remove punctuation and special characters using regex
    cleaned_text = re.sub(pattern, '', text)
    
    return cleaned_text

# Apply the remove_special_characters function to the 'text' column
new_df['text'] = new_df['text'].apply(remove_special_characters)

# Display the updated DataFrame
new_df
```

```python
new_df.to_csv('csv/careers_double.csv',index=False)
```

```python
import random

# Assuming 'new_df' is the DataFrame containing 'Job Title' and 'text'

female = ['daughter', 'niece', 'mom', 'grandma', 'sister', 'aunt', 'wife', 'girlfriend']
male = ['son', 'nephew', 'dad', 'grandpa', 'brother', 'uncle', 'husband', 'boyfriend']

# Sentence templates for females and males
templates = [
    'My [gender] worked as [article] [occupation].',
    'The [gender] had a job as [article] [occupation].',
    'The [gender] started working as [article] [occupation].',
    'My [gender] had a part-time job as [article] [occupation].',
    'The [gender] recently became [article] [occupation].',
    'I know a [gender] who works as [article] [occupation].',
    'The [gender] is pursuing a career as [article] [occupation].',
    'My [gender] has been working as [article] [occupation] for years.',
    'The [gender] is passionate about [article] [occupation] work.',
    'I met a [gender] who is [article] [occupation] by profession.'
]

# Create an empty list to store the generated sentences
sentences = []

# Generate sentences for all job titles
for occupation in new_df['Job Title']:
    female_sentence_template = random.choice(templates)
    article = 'an' if occupation[0].lower() in ['a', 'e', 'i', 'o', 'u'] else 'a'
    female_sentence = female_sentence_template.replace('[gender]', random.choice(female)).replace('[article]', article).replace('[occupation]', occupation)
    
    male_sentence_template = random.choice(templates)
    article = 'an' if occupation[0].lower() in ['a', 'e', 'i', 'o', 'u'] else 'a'
    male_sentence = male_sentence_template.replace('[gender]', random.choice(male)).replace('[article]', article).replace('[occupation]', occupation)
    
    sentences.extend([female_sentence, male_sentence])

# Create the 'ddd' DataFrame
ddd = pd.DataFrame({'prompt': sentences, 'completion': new_df['text'].repeat(2)})

# Display the 'ddd' DataFrame
ddd
```

```python
ddd.to_csv('csv/prompt_data.csv', index=False)
```
